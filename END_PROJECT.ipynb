```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/Certificates/ABC Company.xlsx - Sheet1.csv")
(df.head())
```

```python
(df.info())
(df.describe())
(df.isnull().sum())
```

```python
np.random.seed(10)
df['Height']=np.random.randint(150,181,size=len(df))
df['Height']
```

```python
team_distribution=df['Team'].value_counts()
print(team_distribution)

sns.countplot(data=df,x='Team')
plt.title('Employee Distribution across Teams')
plt.xlabel('Team')
plt.ylabel('No.of Employees')
plt.show()
```

```python
team_percentage=(df['Team'].value_counts()/len(df))*100
print(team_percentage)



```

```python
position_count=df['Position'].value_counts()
print(position_count)


sns.countplot(data=df,x='Position')
plt.title('Employees By Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()
```

```python
bins=[20,25,30,35,40,45,50]
labels=['20-25','26-30','31-35','36-40','41-45','46-50']
df['Age_Group']=pd.cut(df['Age'],bins=bins,labels=labels)
print(df['Age_Group'].value_counts())

sns.countplot(data=df,x='Age_Group')
plt.title('Predominant Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
team_salary=df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
print(team_salary)

plt.bar(team_salary.index,team_salary.values)
plt.title('Total Salary by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.show()
```

```python
position_salary=df.groupby('Position')['Salary'].sum().sort_values(ascending=False)
print(position_salary)

plt.bar(position_salary.index,position_salary.values)
plt.title('Total Salary by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.show()

```

```python
correlation=df['Age'].corr(df['Salary'])
print(correlation)
```

```python
sns.scatterplot(data=df,x='Age',y='Salary')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

Data Story

From the Team Distribution Analysis,it is observed that employees are not equally distributed.New Orleans Pelicans have the highest team percentage.The Age group Analysis shows most emploees fall in the age 20-25,and 26-30.Cleveland Cavaliers  have the highest salary by Team.The position wise Salary shows Centers have the highest Salary expenditure.
